```python
import pandas as pd

df = pd.read_excel("./kaggle_default.xls", sheet_name="Data")
cols = df.iloc[0,:].values
df.columns = cols
df.drop(["ID"], axis=1, inplace=True)
df.drop(index=0, axis=0, inplace=True)
df.rename(columns={"PAY_0":"PAY_1",
                    "default payment next month":"default"}, inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
```

```python
print(train['default'].value_counts())
train
```

```python
print(test['default'].value_counts())
test
```

```python
train.reset_index(drop=True, inplace=True)
train
```

```python
test.reset_index(drop=True, inplace=True)
test
```

```python
train.to_csv('./GDSC_train.csv', index=False)
```
